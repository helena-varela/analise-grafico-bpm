**Análise de Saúde Cardíaca em Python**

```python
import random
import matplotlib.pyplot as plt

nome = input("Nome do paciente: ")
maxima_freq = int(input("A frequência cardíaca máxima esperada (bpm): "))

tempo = [1, 2, 3, 4, 5, 6, 7]
batimentos = []
# Gera 7 valores simulados entre 50 e 120 bpm (representando uma semana)
for i in range(7):
  aleatorio = random.randint(50, 120)
  batimentos.append(aleatorio)

# Funções
def calcular_media(batimentos):
  """Recebe uma lista e retorna a média aritmética."""
  return sum(batimentos)/7

def dias_acima_limite(batimentos, max):
  """Retorna quantos valores excedem o limite informado."""
  qnt = 0
  for i in batimentos:
    if i > max:
      qnt += 1
  return qnt

media = calcular_media(batimentos)
passou_limite = dias_acima_limite(batimentos, maxima_freq)

print("Relatório:")
print(f"- Frequências cardíacas: {batimentos}")
print(f"- Média semanal: {media:.2f}")
print(f"- Dias acima de {maxima_freq}: {passou_limite}\n")
```

```python
# Gerando um Gráfico
plt.plot(tempo, batimentos, marker='o')
plt.title("Análise de BPM ao Longo de uma Semena")
plt.xlabel("Dias")
plt.ylabel("Batimentos por Minuto (BPM)")
plt.grid(True)
plt.show()
```
