**Importações de bibliotecas**

```python
import random
import matplotlib.pyplot as plt
```

**Funções**

```python
def calcular_media(batimentos):
  """Recebe uma lista e retorna a média aritmética."""
  return sum(batimentos)/7

def dias_acima_limite(batimentos, max):
  """Retorna quantos valores excedem o limite informado."""
  qnt = 0
  for i in batimentos:
    if i > max:
      qnt += 1
  return qnt
```

**Código**

```python
# Dados do paciente
nome = input("Nome do paciente: ")
maxima_freq = float(input("A frequência cardíaca máxima esperada (bpm): "))

# Verificação da frequência cardíaca ser um número positivo
if maxima_freq < 0:
  print("A frequência cardíaca máxima esperada precisa ser positiva. Digite novamente.")
else:
  tempo = [1, 2, 3, 4, 5, 6, 7]
  batimentos = []

  semana = {
      "Domingo": 0,
      "Segunda-Feira": 0,
      "Terça-Feira": 0,
      "Quarta-Feira": 0,
      "Quinta-Feira": 0,
      "Sexta-Feira": 0,
      "Sábado": 0,
  }

  dia = list(semana.keys())

  # Gera 7 valores simulados entre 50 e 120 bpm (representando uma semana)
  for i in range(7):
    aleatorio = random.randint(50, 120)
    batimentos.append(aleatorio)
    semana[dia[i]] = aleatorio

  media = calcular_media(batimentos)
  passou_limite = dias_acima_limite(batimentos, maxima_freq)

  # Emissão de Relatório
  print("\nRELATÓRIO:")
  print(f"\n- Frequências cardíacas: {batimentos}")
  print(f"- Média semanal: {media:.2f}")
  print(f"- Dias acima de {round(maxima_freq)}: {passou_limite}")
  if passou_limite > 3:
    print("- Alerta: Consulte um médico!")
  print("- Dias da semana e BPMs ")
  for dia, valor in semana.items():
    print(f'  {dia}: {valor}')
```

**Gerando um Gráfico**

```python
plt.plot(tempo, batimentos, marker='o')
plt.title("Análise de BPM ao Longo de uma Semena")
plt.xlabel("Dias")
plt.ylabel("Batimentos por Minuto (BPM)")
plt.grid(True)
plt.show()
```
